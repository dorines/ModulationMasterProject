```python
# Plot variation in amplitude vs relative/percent error in amplitude/phase/frequency 
# using mean_found_amplitude/phase/frequency_array, stdv_found_amplitude/phase/frequency_array 
# from Sim_radioactive_decay_NO24h_FFT_loop_varying_amplitude

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math
from numpy import *
%matplotlib inline

matplotlib.rc('font', size=18)                   # Use big fonts...
plt.rcParams['figure.figsize'] = (12.0, 10.0)    # ... and big plots

A0 = 37 

# Varying amplitude in percentages.

# input_amplitude_array_annual_mod_005_01 = np.loadtxt("annual_mod_input_amplitude_CosFit_005_01")
# amplitude_percentage_annual_mod_005_01 = 100*input_amplitude_array_annual_mod_005_01/A0

annual_mod_input_amplitude_CosFit = np.loadtxt("annual_mod_input_amplitude_expANDcosFitF")
amplitude_percentage_annual_mod = 100*annual_mod_input_amplitude_CosFit/A0
```

```python
# 500 days exp and cos together:
relative_error_found_amplitude_annual_mod_500_expANDcosFit = np.loadtxt("relative_error_found_amplitude_annual_mod_500_expANDcosFit_500rF")
relative_error_found_frequency_annual_mod_500_expANDcosFit = np.loadtxt("relative_error_found_frequency_annual_mod_500_expANDcosFit_500rF")
relative_error_found_phase_annual_mod_500_expANDcosFit = np.loadtxt("relative_error_found_phase_annual_mod_500_expANDcosFit_500rF")
```

```python
mean_found_amplitude_expANDcosFit_annual_mod_500 = np.loadtxt("mean_found_amplitude_expANDcosFit_annual_mod_500_500rF")
stdv_found_amplitude_expANDcosFit_annual_mod_500 = np.loadtxt("stdv_found_amplitude_expANDcosFit_annual_mod_500_500rF")
mean_found_frequency_expANDcosFit_annual_mod_500 = np.loadtxt("mean_found_frequency_expANDcosFit_annual_mod_500_500rF")
stdv_found_frequency_expANDcosFit_annual_mod_500 = np.loadtxt("stdv_found_frequency_expANDcosFit_annual_mod_500_500rF")
mean_found_phase_expANDcosFit_annual_mod_500 = np.loadtxt("mean_found_phase_expANDcosFit_annual_mod_500_500rF")
stdv_found_phase_expANDcosFit_annual_mod_500 = np.loadtxt("stdv_found_phase_expANDcosFit_annual_mod_500_500rF")
```

```python
# Same plots as above, but compare with and without exp and cos together

figAmpl = plt.figure()
plt.title('Sensitivity research - annual modulation - exp and cos fit', fontsize=20)

# Exponent + Cosine fit:
plt.errorbar(annual_mod_input_amplitude_CosFit, mean_found_amplitude_expANDcosFit_annual_mod_500, stdv_found_amplitude_expANDcosFit_annual_mod_500, label="Found amplitude - 500 days - 500 runs")
#plt.plot(annual_mod_input_amplitude_CosFit, mean_found_amplitude_expANDcosFit_annual_mod_500, 'o')

plt.plot(annual_mod_input_amplitude_CosFit, annual_mod_input_amplitude_CosFit, 'k', label="Input amplitude")
plt.xlabel('input amplitude[Hz]')
plt.xlim(min(annual_mod_input_amplitude_CosFit), max(annual_mod_input_amplitude_CosFit))
#plt.xscale('log')
plt.ylabel('found amplitude [Hz]')
#plt.ylim(-0.05, .4)
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
plt.savefig('input_yearly_found_with_errors_amplitude.png', format='png', dpi=figAmpl.dpi)
plt.show()
```

```python
plt.title('                          Sensitivity research with input value - Frequency - 500 runs')

frequency_input = 1./(365.25*24*60*60.0)
input_frequency_array = np.ones(len(annual_mod_input_amplitude_CosFit))*frequency_input

#plt.errorbar(amplitude_percentage_annual_mod, mean_found_frequency_CosFit_annual_mod_500, stdv_found_frequency_CosFit_annual_mod_500, label="Found frequency with errors - 500 days")
plt.errorbar(amplitude_percentage_annual_mod, mean_found_frequency_expANDcosFit_annual_mod_500, stdv_found_frequency_expANDcosFit_annual_mod_500, label="Found frequency with errors - 500 days with expANDcos together")

plt.plot(amplitude_percentage_annual_mod,input_frequency_array, 'k', label="Input frequency")
plt.xlabel('input amplitude[%]')
#plt.xlim(0.000005*A0/100, 0.2)
#plt.yscale('log')
plt.ylabel('found freqeuncy [Hz]')
plt.ylim(2.7*10**(-8), 3.5*10**(-8))
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
#plt.savefig("input_yearly_found_with_errors_frequency.png", format='png', dpi = figFreq.dpi) 
plt.show()
```

```python
plt.title('Sensitivity research with input value - Phase - annual mod - Cos Fit - 300 runs')

phase_in_Degrees = 40.
phase_input = np.deg2rad(phase_in_Degrees)
input_phase_array = np.ones(len(annual_mod_input_amplitude_CosFit))*phase_input

#plt.errorbar(amplitude_percentage_annual_mod, mean_found_phase_CosFit_annual_mod_500, stdv_found_phase_CosFit_annual_mod_500, label="Found phase with errors - 500 days")
plt.errorbar(amplitude_percentage_annual_mod, mean_found_phase_expANDcosFit_annual_mod_500, stdv_found_phase_expANDcosFit_annual_mod_500, label="Found phase with errors - 500 days with expANDcos together")

plt.plot(amplitude_percentage_annual_mod, input_phase_array, 'k', label="Input phase")
plt.xlabel('input amplitude[%]')
#plt.xlim(0.000005*A0/100, (0.2))
#plt.xlim(0.1, (0.2))
plt.ylabel('found phase')
plt.ylim(-0.1, 1.2)
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
#plt.savefig("input_yearly_found_with_errors_phase.png", format ='png', dpi = 300) 
plt.show()
```
