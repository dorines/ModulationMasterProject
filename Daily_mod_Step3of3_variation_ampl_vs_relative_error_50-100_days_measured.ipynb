```python
#plt.title('Sensitivity research - Amplitude', fontsize=15)

plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_amplitude_array_NO24h_50_V2, stdv_found_amplitude_array_NO24h_50_V2, label="Found amplitude with errors - 50 days")
#plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_amplitude_array_NO24h_100_V2, stdv_found_amplitude_array_NO24h_100_V2, label="Found amplitude with errors - 100 days")
plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_amplitude_array_NO24h_150_V2, stdv_found_amplitude_array_NO24h_150_V2, label="Found amplitude with errors - 150 days")

plt.plot(amplitude_percentage_NO24h_V2,input_amplitude_array_NO24h_V2, 'k', linewidth = 2.5, label="Input amplitude")
plt.xlabel('input amplitude[%]', fontsize = 28)
#plt.xlim(0.000005*A0/100, 0.2)
plt.xlim(0.0, 0.1)
#plt.xscale('log')
plt.ylabel('found amplitude [Hz]', fontsize = 28)
plt.ylim(0.005, 0.06)
plt.legend(bbox_to_anchor=(1.0, 1), loc=1, borderaxespad=0., fontsize = 28)
plt.savefig("input_found_with_errors_amplitude.pdf", format='pdf', dpi=300)
plt.savefig("input_found_with_errors_amplitudePNG.png", format='png', dpi=300)
plt.show()
```

```python
#plt.title('Sensitivity research - Phase', fontsize=15)

phase_in_Degrees = 40.
phase_input = np.deg2rad(phase_in_Degrees)
input_phase_array = np.ones(len(input_amplitude_array_NO24h_V2))*phase_input

plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_phase_array_NO24h_50_V2, stdv_found_phase_array_NO24h_50_V2, label="Found phase with errors - 50 days")
#plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_phase_array_NO24h_100_V2, stdv_found_phase_array_NO24h_100_V2, label="Found phase with errors - 100 days")
plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_phase_array_NO24h_150_V2, stdv_found_phase_array_NO24h_150_V2, label="Found phase with errors - 150 days")

plt.plot(amplitude_percentage_NO24h_V2,input_phase_array, 'k', linewidth = 2.5, label="Input phase")
plt.xlabel('input amplitude[%]', fontsize = 28)
plt.xlim(0.000005*A0/100, (0.2))
#plt.xlim(0.1, (0.2))
plt.ylabel('found phase', fontsize = 28)
#plt.ylim(0., 0.03)
plt.legend(bbox_to_anchor=(1.0, 1), loc=1, borderaxespad=0., fontsize = 28)
#plt.savefig("input_found_with_errors_phase.pdf", format='pdf', dpi=300) 
plt.savefig("input_found_with_errors_phasePNG.png", format='png', dpi=300) 
plt.show()
```

```python
#plt.title('Sensitivity research - Frequency', fontsize=15)

frequency_input = 1./(24*60*60.0)
input_frequency_array = np.ones(len(input_amplitude_array_NO24h_V2))*frequency_input

plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_frequency_array_NO24h_50_V2, stdv_found_frequency_array_NO24h_50_V2, label="Found frequency with errors - 50 days")
#plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_frequency_array_NO24h_100_V2, stdv_found_frequency_array_NO24h_100_V2, label="Found frequency with errors - 150 days")
plt.errorbar(amplitude_percentage_NO24h_V2, mean_found_frequency_array_NO24h_150_V2, stdv_found_frequency_array_NO24h_150_V2, label="Found frequency with errors - 150 days")

plt.plot(amplitude_percentage_NO24h_V2,input_frequency_array, 'k', linewidth = 2.5, label="Input frequency")
plt.xlabel('input amplitude[%]', fontsize = 28)
plt.xlim(0.000005*A0/100, 0.2)
#plt.xscale('log')
plt.ylabel('found freqeuncy [Hz]', fontsize = 28)
plt.ylim(-0.002, 0.009)
#plt.yscale('log')
plt.legend(bbox_to_anchor=(1.0, 1), loc=1, borderaxespad=0., fontsize = 28)
#plt.savefig("input_found_with_errors_frequency.pdf", format='pdf', dpi=300) 
plt.savefig("input_found_with_errors_frequencyPNG.png", format='png', dpi=300) 
plt.show()
```

```python
# Plot Sensitivity research - Amplitude - 24h for amplitude is 0.01-1.9%

plt.title('Sensitivity research - Amplitude', fontsize=15)
plt.plot(amplitude_percentage_NO24h_V2, relative_error_found_amplitude_50_NO24h_V2, 'g', label="RSD in found amplitude - 50 days")
plt.plot(amplitude_percentage_NO24h_V2, relative_error_found_amplitude_100_NO24h_V2, 'b', label="RSD in found amplitude - 100 days")
plt.plot(amplitude_percentage_NO24h_V2, relative_error_found_amplitude_150_NO24h_V2, 'r', label="RSD in found amplitude - 150 days")
#plt.plot(amplitude_percentage_NO24h_V2, relative_error_found_amplitude_200_NO24h_V2, 'k', label="RSD in found amplitude - 200 days")
#plt.plot(amplitude_percentage_NO24h_V2, relative_error_found_amplitude_300_NO24h_V2, 'c', label="RSD in found amplitude - 300 days")
plt.xlabel('amplitude[%]', fontsize=12)
plt.xlim(min(amplitude_percentage_NO24h_V2), max(amplitude_percentage_NO24h_V2))
plt.xscale('log')
plt.ylabel('relative standard deviation (RSD) [%]', fontsize=12)
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
plt.show()
print("Half life time: 5.27 years, frequency input = 1./(24*60*60.0) = 1 modulation per day and phase input = 40 degrees")
print("amplitude input = percentage*A0 and RSD = (standard_deviation_found_ampl/freq/phase_array)/(mean_found_ampl/freq/phase_array)")
```
